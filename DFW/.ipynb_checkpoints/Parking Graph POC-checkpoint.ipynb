Setting garage fill levels. These will be used in our price calculation.

```python
gar=['A_garage_1','A_garage_2','B_garage_1','B_garage_2',
 'C_garage_1','C_garage_2','D_garage_1','D_garage_2',
 'E_garage_1','E_garage_2','Express_North','Express_South']
fills=[0.85,0.8,0.75,0.7,
       0.75,0.70,0.9,0.5,
      0.75,0.7,0.3,0.3]
```

```python
nx.shortest_path(drg,source='North_Entrance',target='A_garage_1',weight='duration')
```

Function that shows the shortest paths and costs from all different parking options. Prices are adjusted by fill level and bounded by a coefficient of the user's choising (i.e. prices can never drop below 0.7 of original rate).

```python
def shortlendfw(entry,gate,rate=25,lowbound=0.7,days=3):
    patdrg=[nx.shortest_path(drg,entry,i,weight='duration') for i in gar]
    lendrg=[nx.shortest_path_length(drg,entry,i,weight='duration') for i in gar]
    pating=[nx.shortest_path(ing,i,gate,weight='duration') for i in gar]
    lening=[nx.shortest_path_length(ing,i,gate,weight='duration') for i in gar]
    pathdat=OrderedDict([('garage',gar),
                        ('path to garage',patdrg),
                        ('length to garage',lendrg),
                        ('path from garage',pating),
                        ('length from garage',lening)])
    pathout=pd.DataFrame.from_dict(pathdat)
    pathout['total length']=pathout['length to garage']+pathout['length from garage']
    pathout['fill']=fills
    pathout['adjusted price']=lowbound*rate+pathout['fill']*(1-lowbound)*rate
    pathout['total']=pathout['adjusted price']*days
    #pathg=dg.subgraph(pathout.head(1)[])
    return(pathout.sort_values('total length'))
```

```python
x=shortlendfw('South_Entrance','E_gates_2',days=5)
pathg=dg.subgraph(list(x.head(1)['path to garage'])[0]+
                 list(x.head(1)['path from garage'])[0])
nx.draw_networkx(pathg,color=colors)
x
```

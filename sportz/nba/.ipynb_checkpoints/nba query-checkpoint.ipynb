```python
import nba_py
from nba_py import player
from nba_py import constants
import requests
import json
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
years=np.arange(1985,2019)
tabledat=[]

for year in years:
    path='https://www.basketball-reference.com/leagues/NBA_'+str(year)+'_advanced.html'
    soup=BeautifulSoup(requests.get(path).text,'html.parser')
    tab=soup.find('table')
    if year==1985:
        headers=tab.find_all('th')
        #header=headers[0].text.strip()
        headers=[header.text.strip() for header in headers]
        headers=headers[1:29]
        headers=[header for header in headers if len(header)>0]
        headers.append('Year')
        print(headers)

    rows=tab.find_all('tr')
    for row in rows:
        cols=row.find_all('td')
        cols=[ele.text.strip() for ele in cols]
        rowdat=[ele for ele in cols if ele]
        if len(rowdat)>0:
            rowdat.append(year)
            tabledat.append(rowdat)
    
advdat=pd.DataFrame(tabledat,columns=headers)
```

```python
filename='nbadraft.json'
if filename:
    with open(filename,'r') as f:
        datastore=json.load(f)
```

```python
dat=datastore['resultSets'][0]['rowSet']
cols=datastore['resultSets'][0]['headers']
draftdat=pd.DataFrame(dat,columns=cols)
draftdat.info()
```

```python
draftdat.head()
```

```python
draftdat.SEASON=pd.to_numeric(draftdat.SEASON)
lotdat=draftdat[(draftdat.SEASON>=1985)&(draftdat.ROUND_NUMBER<=2)]
```

```python
lotdat.tail()
```

```python
lotdat['Draft_Year']=lotdat['SEASON']
lotdat['EXPERIENCE']=0
lotdat.tail()
```

```python
lotdat2=pd.DataFrame()

for i in range(5):
    lotdatadd=lotdat.copy(deep=True)
    inc=1+i
    lotdatadd['SEASON']=lotdatadd['SEASON']+inc
    lotdatadd['EXPERIENCE']=lotdatadd['EXPERIENCE']+inc
    lotdat2=lotdat2.append(lotdatadd)
lotdat2=lotdat2[lotdat2.SEASON<2019].sort_values(["Draft_Year","SEASON","OVERALL_PICK"],ascending=[False,True,True])
```
